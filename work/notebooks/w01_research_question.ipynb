## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

Based on a quick analysis of the starter CSV data, here are the real numbers that prove this lane is worth investigating:

The overall median word count for content in this dataset is 2792 words.

When broken down by trend direction, 'down' pages have a median word count of 2909 words, while 'up' pages have a median of 2848 words.

This means declining pages are actually slightly longer on average than growing ones, clearly indicating that word count alone is not the magical growth lever it is rumored to be.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
# The real numbers backing up the text above
overall_median = df["word_count"].median()
by_trend = df.groupby("trend_direction")["word_count"].median()

print(f"Overall Median Word Count: {overall_median}")
print("\nMedian Word Count by Trend Direction:")
print(by_trend.to_string())
```
